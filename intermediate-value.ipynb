```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

x = sp.Symbol('x')
# This is the plotting function that is repeatedly used, so define it and pass in f, a, b, title from code below.
def plot_function(f, a, b, title, y_target=None):
    x_vals = np.linspace(a, b, 400)
    y_vals = f(x_vals)
    plt.figure(figsize=(8, 4))
    plt.plot(x_vals, y_vals, label='f(x)', color='blue')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(a, color='gray', linestyle='--')
    plt.axvline(b, color='gray', linestyle='--')
    if y_target is not None:
        plt.axhline(y_target, color='red', linestyle='--', label=f'y = {y_target}')
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

Intermediate Value Theorem
- the function $f$ is continuous on the interval $(a,b)$
- $f(a) \ne f(b)$
- $c$ is in the interval $(a,b)$
- $f(c) = N$ is between $f(a)$ and $f(b)$

Let $f(x) = x^3 - x - 2$.

Which of the following intervals must contain a solution to the equation $f(x) = 0$, based on the Intermediate Value Theorem?

    A. (0, 1)  
    B. (1, 2)  
    C. (2, 3)  
    D. (3, 4)

```python
# A. Open interval (0, 1)
a = 0
b = 1
f = x**3 - x - 2
flambda = sp.lambdify(x, f, 'numpy')
plot_function(flambda, a, b, " f(x) = xÂ³ - x - 2", y_target=0)
fa = f.subs(x, a).evalf()
fb = f.subs(x, b).evalf()
fc = 0
fa, fc, fb
```
